# Data Processing Pipeline

```python
import pandas as pd 
import numpy as np 
import bamboolib as bam 
bam.enable()
DATA_PATH = 'intermediate/Pivoted_Data_10Cities.xlsx'
```

## Reading Dataset

```python
data = pd.read_excel(DATA_PATH)
```

```python
data
```

```python
#rename provinsi into "tanggal"
data = data.rename(columns={'Provinsi':'tanggal'})
data['tanggal'] = pd.to_datetime(data['tanggal'])
```

```python
data.dtypes
#seems like all data types are object.Check Unique Values
```

```python
data.columns
```

```python
data['Kota Bandung'].unique()
# there is '-' in Kota Bandung unique val, replace into np,NaN
```

```python
data = data.replace('-',np.NaN)
```

```python
#check dtypes again 

cities = ['Kota Bandung', 'Kota Makassar', 'DKI Jakarta', 'Medan',
       'Surabaya', 'Kota Pekanbaru', 'Kota Batam', 'Kota Palembang',
       'Kota Yogyakarta', 'Kota Balikpapan']
data[cities] = data[cities].astype('float')
```

```python
data.dtypes
```

```python
data = data.rename(columns={'Kota Bandung': 'Bandung', 'Kota Makassar': 'Makassar', 'Kota Pekanbaru': 'Pekanbaru', 'Kota Batam': 'Batam', 'Kota Palembang': 'Palembang', 'Kota Yogyakarta': 'Yogyakarta', 'Kota Balikpapan': 'Balikpapan'})
data
```
