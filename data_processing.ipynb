```python
import pandas as pd 
import numpy as np 

DATA_PATH = 'Compiled_All_Cities.xlsx'
```

```python
data =pd.read_excel(DATA_PATH)
```

```python
data['tanggal'] = pd.to_datetime(data['tanggal'], format='%d/%m/%Y')
```

```python
data = data.rename(columns={'Kota Medan': 'medan', 'Kota Pekanbaru': 'pekanbaru', 'Kota Batam': 'batam', 'Kota Palembang': 'palembang', 'Kota Bandung': 'bandung', 'Kota Jakarta Pusat': 'jakarta', 'Kota Yogyakarta': 'yogyakarta', 'Kota Surabaya': 'surabaya', 'Kota Balikpapan': 'balikpapan'})
data
```

```python
idx = pd.date_range("2020-01-01", "2022-11-18")
# set index
data = data.set_index("tanggal")
data = data.reindex(idx, fill_value=np.NaN)
```
